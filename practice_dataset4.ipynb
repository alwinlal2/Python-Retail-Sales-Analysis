```python
#correlation check
```

```python
df.corr(numeric_only = True)
corr_matrix = df.corr(numeric_only = True)
sns.heatmap(corr_matrix , annot =True, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(5,2))
plt.hist(df['Total Sales (INR)'], bins=20, edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('rupees')
plt.ylabel('frequency')
plt.grid(alpha=0.4)
plt.show()
```

```python
sales_category = df.groupby('Category')['Total Sales (INR)'].sum().sort_values(ascending = False)
plt.figure(figsize=(5,2))
sales_category.plot(kind = 'bar', color = 'pink' , edgecolor = 'black')
plt.title('Sales by category')
plt.xlabel('category')
plt.ylabel('rupees')
plt.grid(alpha=0.4)
plt.show()
```

```python
monthly_trend = df.groupby('month')['Total Sales (INR)'].sum()

plt.figure(figsize=(5,2))
monthly_trend.plot(marker = 'o', linestyle ='-' , color = 'red')
plt.title('monthly revenue trend')
plt.xlabel('month-year')
plt.ylabel('sales')
plt.grid(alpha=0.2)
plt.show()
```

```python
payment_method = df.groupby('Payment Method')['Total Sales (INR)'].sum().sort_values(ascending = False)

plt.figure(figsize=(5,3))
payment_method.plot(kind='pie', autopct = '%1.1f%%')
plt.title('sales by payment method')
plt.ylabel('')
plt.show()
```
